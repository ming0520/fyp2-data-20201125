```python
#funtion use to beautify json file
def beautify_json(input_file,output_file,verbose=False):
    file = open(input_file)
    mydata = json.load(file)
    if(verbose):
        print(json.dumps(mydata, indent=4, sort_keys=True))
    with open(f'{output_file}', 'w') as f:
        json.dump(mydata, f,indent=4, sort_keys=True)
```

```python
#Load all the file
input_path = 'GentleAligned'
output_path = 'GentleAlignedBeautify'
files = glob.glob(f'{input_path}/*.json')
files_name = []
for file in files:
#     print(file.split('\\')[1].split('.')[0])
    files_name.append(file.split('\\')[1].split('.')[0])

# input_files = []
# for file_name in files_name:
#     relative_file = f'./{input_path}/' + file_name + '.json'
#     input_files.append(relative_file)
    
# output_files = []
# for file_name in files_name:
#     relative_file = f'./{output_path}/' + file_name + '_Beautified'+ '.json'
#     output_files.append(relative_file)
```

```python
for file_name in files_name:
    input_file = f'./{input_path}/' + file_name + '.json'
    output_file = f'./{output_path}/' + file_name + '.json'
    beautify_json(input_file,output_file)
#     print(input_file, output_file)
```
